```python
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns
import datetime as dt
import numpy as np
```

```python
covid= pd.read_csv('covid_19_india.csv', parse_dates=['Date'], dayfirst=True)
```

```python
covid.head()
```

```python
covid.tail()
```

```python
covid= covid[['Date', 'State/UnionTerritory', 'Cured', 'Deaths', 'Confirmed']]
covid
```

```python
covid.columns= ['date','state','cured','death','confirmed']
covid.head()
```

```python
covid.tail()
```

```python
today= covid[covid.date=='2020-04-11']
today
```

```python
#Sorting data w.r.t number of confirmed cases
max_confirmed_cases= today.sort_values(by='confirmed',ascending=False)
max_confirmed_cases
```

```python
top_states_confirmed= max_confirmed_cases[0:5]
top_states_confirmed
```

```python
#Making bar-plot for states with top confirmed cases
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x="state",y="confirmed",data=top_states_confirmed)
plt.show()
```

```python
#Sorting data w.r.t number of cured cases
max_cured_cases= today.sort_values(by='cured',ascending=False)
max_cured_cases
```

```python
top_states_cured= max_cured_cases[0:5]
top_states_cured
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x="state",y="cured",data=top_states_cured)
plt.show()
```

```python
max_death_cases= today.sort_values(by='death',ascending=False)
max_death_cases
```

```python
top_states_death= max_death_cases[0:5]
top_states_death
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x="state",y="death",data=top_states_death)
plt.show()
```
